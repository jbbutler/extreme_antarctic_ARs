```python
import numpy as np
import json
import os
```

## Snow Model

### Round 1

```python
hyperparam_dict = {}

hyperparam_dict['etas'] = [0.001, 0.01, 0.05, 0.1, 0.3]
hyperparam_dict['gammas'] = [0, 0.0001, 0.001, 0.01, 1, 10, 100]
hyperparam_dict['max_depth'] = [1, 3, 5, 7, 10, 15]
hyperparam_dict['lambdas'] = [0, 0.001, 0.01, 0.1, 1, 10, 100]
hyperparam_dict['min_child_weights'] = [0, 1, 5, 10, 100, 1000]
hyperparam_dict['subsample_fracs'] = [0.1, 0.3, 0.5, 0.7, 1]
hyperparam_dict['nrounds'] = 20000
hyperparam_dict['early_stopping_rounds'] = 20
hyperparam_dict['tree_method'] = 'exact'
hyperparam_dict['booster'] = 'gbtree'
```

```python
with open('snow_round1.json', 'w') as json_file:
    json.dump(hyperparam_dict, json_file, indent=4)
```
