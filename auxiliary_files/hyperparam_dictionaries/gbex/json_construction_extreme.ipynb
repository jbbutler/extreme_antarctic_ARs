# JSON Construction for Extreme Quantile Regression

Helper notebook to construct json representations of python dictionaries, with all of the hyperparameters to test via CV in a certain round of hyperparam searching. This is specifically to perform model selection for a quantile regression model to predict extreme quantiles of conditional outcome distributions.

```python
import numpy as np
import json
import os
curwd = os.getcwd()
```

```python
hyperparam_dict = {}
# add the max number of trees as a hyperparam
hyperparam_dict['depth_sigma'] = [0, 1, 2, 3]
hyperparam_dict['depth_gamma'] = [0, 1, 2, 3]
hyperparam_dict['lambda_scale'] = [0.1, 0.05, 0.01, 0.001]
hyperparam_dict['lambda_ratio'] = [0.5, 1, 2]
hyperparam_dict['min_leaf_sigma'] = [3, 5, 7, 10]
hyperparam_dict['min_leaf_gamma'] = [3, 5, 7, 10]
hyperparam_dict['sample_frac'] = [0.4, 0.6, 0.8, 1]
hyperparam_dict['max_trees'] = 1000
```

```python
with open(curwd + '/rounds_extreme/snow_extreme_round1_hyperparams.json', 'w') as json_file:
    json.dump(hyperparam_dict, json_file, indent=4)
```
