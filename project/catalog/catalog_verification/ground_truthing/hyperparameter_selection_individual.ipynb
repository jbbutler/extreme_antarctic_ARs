There are several hyperparameters we can vary in the clustering algorithm, including `epsilon_space`, `epsilon_time`, `min_pts`, and `rep_pts`. We will first investigate our default of hyperparameters, demonstrating this set of hyperparameters correctly finds and distinguishes ARs identified in published case studies. We will then show limitations for other choices of hyperparameters, justifying our choices.

Defaults are the following, and in the below, the parameter value listed indicates the one which was changed.
+ `epsilon_space`: 0.5
+ `epsilon_time`: 12
+ `min_pts`: 5
+ `rep_pts`: 10
