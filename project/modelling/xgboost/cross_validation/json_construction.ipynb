# JSON Construction

Helper notebook to construct json representations of python dictionaries, with all of the hyperparameters to test via CV in a certain round of hyperparam searching.

```python
import numpy as np
import json
import os
curwd = os.getcwd()
```

```python
hyperparam_dict = {}

hyperparam_dict['etas'] = [0.01, 0.02, 0.025, 0.05, 0.075]
hyperparam_dict['gammas'] = np.linspace(0.01, 5, 20).tolist()
hyperparam_dict['max_depth'] = np.array([5]).tolist()
hyperparam_dict['lambdas'] = np.linspace(0, 7, 20).tolist()
hyperparam_dict['min_child_weights'] = np.linspace(20, 30, 20).tolist()
hyperparam_dict['nrounds'] = 40000
hyperparam_dict['early_stopping_rounds'] = 20
hyperparam_dict['tree_method'] = 'exact'
hyperparam_dict['booster'] = 'gbtree'
```

```python
with open(curwd + '/rounds/temp_round4_hyperparams.json', 'w') as json_file:
    json.dump(hyperparam_dict, json_file, indent=4)
```
