```python
import numpy as np
import json
import os
curwd = os.getcwd()
```

```python
hyperparam_dict = {}

hyperparam_dict['etas'] = np.array([0.001, 0.005, 0.01]).tolist()
hyperparam_dict['gammas'] = np.array([0.0005, 0.001, 0.005]).tolist()
hyperparam_dict['max_depth'] = np.array([3, 7, 10]).tolist()
hyperparam_dict['lambdas'] = np.array([0, 0.3, 0.6]).tolist()
hyperparam_dict['min_child_weights'] = np.array([0.0005, 0.001, 0.005]).tolist()
hyperparam_dict['nrounds'] = 10
hyperparam_dict['early_stopping_rounds'] = 1
hyperparam_dict['tree_method'] = 'exact'
hyperparam_dict['booster'] = 'gbtree'
```

```python
with open(curwd + '/rounds/test_hyper.json', 'w') as json_file:
    json.dump(hyperparam_dict, json_file, indent=4)
```
