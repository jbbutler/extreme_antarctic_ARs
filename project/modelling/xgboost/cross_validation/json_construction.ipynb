```python
import numpy as np
import json
```

```python
hyperparam_dict = {}

hyperparam_dict['etas'] = np.array([0.001, 0.005, 0.01, 0.05, 0.1, 0.3, 0.5, 0.7, 1]).tolist()
hyperparam_dict['gammas'] = np.array([0.0005, 0.001, 0.005, 0.01, 0.05]).tolist()
hyperparam_dict['max_depth'] = np.array([3, 7, 10, 15, 30]).tolist()
hyperparam_dict['lambdas'] = np.array([0, 0.3, 0.7, 1, 5]).tolist()
hyperparam_dict['min_child_weights'] = np.array([0.0005, 0.001, 0.005, 0.01, 0.05]).tolist()
hyperparam_dict['nrounds'] = 500
hyperparam_dict['early_stopping_rounds'] = 10
hyperparam_dict['tree_method'] = 'exact'
hyperparam_dict['booster'] = 'gbtree'
```

```python
with open('/rounds/snow_round1_hyperparams.json', 'w') as json_file:
    json.dump(hyperparam_dict, json_file, indent=4)
```
