# Test xgb CV functionality

```python
import xgboost as xgb
from pathlib import Path
import numpy as np
from sklearn.model_selection import KFold
import os
import pandas as pd
```

```python
load_path = Path(os.getcwd()).parents[2]/Path('dataset/datasets/model_ready/train.csv')
```

```python
x_cols = ['max_ocean_SLP_gradient', 'max_landfalling_v850hPa', 'avg_landfalling_minomega', 'max_IWV_ais', 'cumulative_landfalling_area', 'max_south_extent']
y_col = 'cumulative_snowfall_ais'
```

```python
train_data = pd.read_csv(load_path, index_col='Label')
X = train_data[x_cols]
y = train_data[y_col]
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=12345)
```

```python
from cv_utils import kfold_cv
```

```python
params = dict(booster='gbtree',
                   eta=0.1,
                   gamma=15,
                   max_depth=10,
                   reg_lambda=10,
                   min_child_weight=0.1,
                   tree_method='exact',
                   objective='reg:squarederror',
                   eval_metric='rmse')
```

```python
res = kfold_cv(params, 1000, 10, kf, X, y)
```

```python
res['model_set'][0].save_model('mod.json')
```
