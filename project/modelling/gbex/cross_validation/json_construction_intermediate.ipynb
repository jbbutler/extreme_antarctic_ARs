# JSON Construction for Intermediate Quantile Regression

Helper notebook to construct json representations of python dictionaries, with all of the hyperparameters to test via CV in a certain round of hyperparam searching. This is specifically to perform model selection for a quantile regression model to predict intermediate quantiles of conditional outcome distributions.

```python
import numpy as np
import json
import os
curwd = os.getcwd()
```

```python
hyperparam_dict = {}

hyperparam_dict['num_trees'] = [10, 100, 200, 500, 750, 1000, 1500, 2000, 5000, 10000]
hyperparam_dict['min_node_size'] = [1, 5, 10, 15, 20, 25, 50, 75, 100, 150, 500]
hyperparam_dict['sample_frac'] = [0.2, 0.4, 0.6, 0.8, 1]
hyperparam_dict['alpha'] = [0, 0.001, 0.005, 0.01, 0.05, 0.1, 0.25, 0.40]
hyperparam_dict['intermediate_quantile'] = 0.8
hyperparam_dict['honesty'] = False
```

```python
with open(curwd + '/rounds_intermediate/temp_intermediate_round1_hyperparams.json', 'w') as json_file:
    json.dump(hyperparam_dict, json_file, indent=4)
```
